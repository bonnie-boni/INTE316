# Question B

 Using PYTHON show how the following is achieved(PRACTICAL)

i. Differentiation

ii. Numerical integration

iii. Curve Fitting

iv. Linear Regression

v. Spline Interpolation

```python
# Diffrentiation

x = sp.symbols('x')
f = x**3 + x**2 + x + 1
f_prime = sp.diff(f, x)
print(f'Diffrentiation : {f_prime}')


# Numerical Intergration

def integrand(x):
    return x**2
result, _ = quad(integrand, 0, 1)
print(f'\nNumerical Intergration : {result}')

# Curve fitting

def func(x, a, b, c):
    return a * np.exp(-b * x) + c

xdata = np.linspace(0, 4, 50)
ydata = func(xdata, 2.5, 1.3, 0.5) + 0.2 * np.random.normal(size=len(xdata))

popt, pcov = curve_fit(func, xdata, ydata)
print(f'\nCurve fitting : {popt}')

# Linear Regression

X = np.array([[1], [2], [3], [4], [5]])
y = np.array([1, 3, 2, 5, 4])

model = LinearRegression().fit(X, y)
print(f'\nLinear Regression : \n\tcoefficients {model.coef_}, \n\tIntercepts{model.intercept_}')

# Spline Interpolation

x = np.array([2.00, 4.25, 5.25, 7.81, 9.20, 10.60])
y = np.array([7.2, 7.1, 6.0, 5.0, 3.5, 5.0])

cs = CubicSpline(x, y)
print(f'\nSpline interpolation : {cs(4.0)}')
```

# Question C

A smart robotic agent with a laser Scanner is doing a quick quality check on holes drilled
in a rectangular plate. The centers of the hole in the plate describes the path the arm
needs to take, and the hole centers are located on a Cartesian coordinate system as shown

X (in) 2.00 4.25 5.25 7.81 9.20 10.60

Y(in) 7.2 7.1 6.0 5.0 3.5 5.0

If the laser scanner is traversing from x=2.00 to x =4.25 in a linear path, what is the value
of y at x= 4.0 using the linear spline formula , show how this problem can be solved using
PYTHON (PRACTICAL)

```python
x1, y1 = 2.00, 7.2
x2, y2 = 4.25, 7.1
x = 4.0

y = y1 + (x - x1) * (y2 - y1) / (x2 - x1)
print(f'at X = 4.0 : \n\t Y = {y}')
```

# Question G

Write a program to show how the trapezoidal rule of integration works in PYTHON 

```python
def trapezoidal_rule(func, a, b, n):
    h = (b - a) / n
    x = np.linspace(a, b, n+1)
    y = func(x)
    return (h/2) * (y[0] + 2 * np.sum(y[1:n]) + y[n])

def f(x):
    return x**2

result = trapezoidal_rule(f, 0, 1, 100)
print(result)
```
